##  Abstract

Manhattan, New York is well known for it's job opportunities. By examining the list of open positions from New York State, specifically in the borough, Manhanhattan, people can quantify which education level is sufficient enough to be able to work less hours per week and still achieve a higher salary than the average listed on the open positions. 

The models and methods needed to find a solution involved steps to clean the data set, find the average of hourly wage and weekly hours for each Education Requirement; implement Multiple Linear Regression; create bar charts and SDE plots; split data into testing and training sets; and create a Decision Tree Classifier. 

At the end of the analysis, the data shows that having an Associate's Degree requires less hours of work per week and higher hourly wage. The second choice would be a Bachelor's Degree, with higher wages, but more hours of work than the positions opened. A position with a Master's Degree requirement has more than the average hours of work per week and average hourly wage, but the average was less than Bachelor's Degree and Associate's Degree. For those with High School Diplomas, the average hours of work per week and average hourly wage were shown as below average.

#### Decision Tree

The Decision Tree model can be used as a predictive model to determine what Education Requirement is the best option in terms of hourly wage and work hours per week. 

```python
drop_cols = ['BoroughManhattan', 'HourlyAnnual', 'EducationRequired']
df3 = df2.drop(drop_cols, axis = 1)
df3
```

#### Creating Dependant and Independant Variables: 
The data frame,'df3' was created in order to make the decision tree categories specific to Average Hours Per Week and Average Wage to prepare for the data for a training and test set as an independent variable, to determine the dependent variable, "Education Required". 70% training and 30% test using a random state as an integer will create same results each time the code is rerun.

```python
# Choose feature columns
feature_cols = df3
X = feature_cols  # Features
y = df2.EducationRequired # targetvariable
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3) 
```

```python
#Create a DT Regression Model.

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

clf = tree.DecisionTreeClassifier()

#Train  the model with training sets
clf.fit(X_train, y_train)

#Predict 
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
clf.get_params()
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
clf.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions, labels=[0,1])
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
fig = plt.figure(figsize=((15,15)))
plot_tree(clf,
            feature_names = df3.columns,
            class_names= ["High School Diploma", "Associate's", "Bachelor's", "Master's"], 
            impurity=False,
            proportion=True,
            filled=True)
fig.savefig('test.png')
```

### 3.4 Results and Analysis of Data

The accuracy for the decision tree is 75%, higher than 50%. The tree is broken down based on whether each Education category is above or below the average of the data. The decision tree determined that any position requiring at the minimum, an Associate's Degree, has a higher average hourly wage in the data set. At the bottom of decision three, the Associates degree is above average for wage and below average for hours per week. Ultimately, position that require a Bachelor's Degree has higher wages and higher average work hours per week.
